```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('focusing-on-mobile-app-or-website.zip')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
g = sns.jointplot(x="Time on Website", y="Yearly Amount Spent", data=df, alpha =0.5)
plt.show()
```

```python
g = sns.jointplot(x="Time on App", y="Yearly Amount Spent", data=df)
plt.show()
```

```python
sns.lmplot(x='Length of Membership',
           y='Yearly Amount Spent',
           data = df,
           scatter_kws={'alpha':0.3})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y=df['Yearly Amount Spent']
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#training the model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
print(cdf)
```

```python
#predictions
```

```python
predictions= lm.predict(x_test)
predictions
```

```python
sns.scatterplot(x=predictions, y=y_test)  # Specify x and y using keyword arguments
plt.xlabel('Predictions')
plt.ylabel('Y Test')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import math
print('MAE:',mean_absolute_error(y_test,predictions))
print('MSE:',mean_squared_error(y_test,predictions))
print('RMSE:',math.sqrt(mean_squared_error(y_test,predictions)))
```

```python
#residuals
```

```python
residuals=y_test-predictions
```

```python
sns.displot(residuals,bins=20,kde=True)
```

```python
import pylab
import scipy.stats as stats
stats.probplot(residuals, dist="norm", plot=pylab)
pylab.show()
```
